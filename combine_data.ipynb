```python
df1 = pd.read_csv("resume_rewrite_df1_200_1331 - processed_df1.csv")
df1.columns = df1.iloc[0]  # set first row as header
df1 = df1[1:]  # drop the first row from data
df1.reset_index(drop=True, inplace=True)
df1.head()
```

```python
df2 = pd.read_csv("resume_df2 (1).csv", encoding='latin1')
df2 = df2[:199]
df2.head()
```

```python
df3 = pd.read_csv("df3_rewrite.csv")
df3.rename(columns={'Unnamed: 0': 'CV', 'Rewritten CV':'Rewritten'}, inplace=True)
df3 = df3.drop('Error Log', axis=1)
# df3.set_index('index', inplace=True)
df3.head()
```

```python
df4 = pd.read_csv("rewrite_df4.csv")
df4 = df4.drop('Model', axis=1)
df4 = df4[:50]
df4.head()
```
