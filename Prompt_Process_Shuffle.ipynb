```python
df1_rewritten.head(5)
```

```python
df_labeled = pd.concat([df1_rewritten, df2_rewritten, df3_rewritten, df4_rewritten, df5_rewritten], ignore_index=True)
```

```python
df_gp = df_labeled.groupby('Prompt').agg({'rewritten_y': 'sum'}).reset_index()
df_gp.head()
```
